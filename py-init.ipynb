# Analyse de données via Pandas

> Cellule 1 importe "pandas" et affiche la version

````python
import pandas as pd
# Affichage de la version
print("Pandas version ", pd.__version__)
````

## Lire le fichier csv

La fonction pd.read_csv() charge un fichier et le transforme en **DataFrame**

> Un *DataFrame* est un tableau de données que Python peut manipuler

```python
datas = pd.read_csv(<filepath>)
```


````python
datas = pd.read_csv("sample_data.csv")

print(type(datas))
````

## Déterminer le nombre de lignes lues, le nombre de colonnes

- **shape** est un **Tuple** au sens Python du terme : une liste *(lignes, colonnes)*
- l'accès aux données d'un Tuple se fait par la notation *[]*

````python
nb_lignes = datas.shape[0]
nb_colonnes = datas.shape[1]
print("Lines number : ", nb_lignes)
print("Columns number : ", nb_colonnes)
print("Complete dimensions : ", datas.shape)
````

## Récupérer les données d'en-tête (les noms des colonnes)

- DataFrame.columns : Retourne le nom des colonnes

## Afficher n lignes pour contrôle

- DataFrame.head(n)

## Connaître le type des colonnes

- DataFrame.dtypes

````python
print("Colonnes")
print(datas.columns)

# Version avec boucle
for col in datas.columns:
    """
        Utilisation de la boucle for... in pour itérer sur une structure Itérable
    """
    print(f"Feature : {col} - Type : {datas[col].dtype}")
    
    if col == "feature_a":
        print(f"Manual mean of feature_a : {datas['feature_a'].mean()}")
    elif col == "target":
        print(f"Partition of target variable : \n{datas['target'].value_counts()}")
    
print("Aperçu des données")
print(datas.head(10))

print("Types des colonnes")
print(datas.dtypes)
````

## Résumé statistique du DataFrame (cumuls, moyennes, écarts-types, ...)

- DataFrame.describe()

````python
datas.describe()
````

## En manuel : pour calculer la moyenne d'une feature

- La fonction "mean()" fait le travail

### Par extension 

- Min : DataFrame["feature"].min()
- Max : DataFrame["feature"].max()

````python
print(f"Manual mean of feature_a : {datas['feature_a'].mean()}")
````

## Sur des chaînes il est impossible de faire des moyennes, des écarts-type...

> On peut par contre utiliser "value_counts()" pour déterminer les valeurs uniques des colonnes

````python
print(f"Partition of target variable : \n{datas['target'].value_counts()}")
````
